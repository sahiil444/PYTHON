```python
# opening a csv file from URL

import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

```python
# usecols parameter is used to specify which columns should be loaded while reading a file into Pandas.
pd.read_csv('aug_train.csv',usecols=['enrollee_id','gender','education_level'])
```

```python
# squeeze parameter The squeeze() method is used to convert a single-column or single-row DataFrame into a Series by reducing its dimensionality.
pd.read_csv('aug_train.csv',usecols=['enrollee_id']).squeeze()
```

```python
# skiprows is used to skip specified rows while reading a file into a DataFrame
pd.read_csv('aug_train.csv',skiprows=[1,5])
```

```python
# nrows is used to specify the number of rows to read from a file.
pd.read_csv('aug_train.csv',nrows=100)
```
