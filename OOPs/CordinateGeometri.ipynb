## Write OOP classes to handle the following scenarios;
#### A user can create and view 2D coordinates
#### A user can find out the distance between 2 coordinates
#### A user can find find the distance of a coordinate from origin
#### A user can check if a point lies on a given line
#### A user can find the distance between a given 2D point and a given line

```python
class Point:

    def __init__(self,x,y):
        self.x_cord = x
        self.y_cord = y

    def __str__(self):
        return f"<{self.x_cord},{self.y_cord}>"

    def distance_dtw_Two_Cordinate(self,other):
        return ((other.x_cord-self.x_cord)**2 + (other.y_cord-self.y_cord)**2)**0.5

    def distance_from_origin(self):
        return (self.x_cord**2 + self.y_cord**2)**0.5
        
class Line:

    def __init__(self,A,B,C):
        self.A = A
        self.B = B
        self.C = C

    def __str__(self):
        return f"{self.A}X + {self.B}Y + {self.C}"

    def point_on_line(line,point):
        if((line.A*point.x_cord + line.B*point.y_cord + line.C)==0):
            return "Point is lies on line"

        else:
            return "Point does not lies on line"

    def sortest_distance_point_line(line,point):
        return abs((line.A*point.x_cord+line.B*point.y_cord+line.C)/(line.A**2+line.B**2)**0.5)
    
    def intersecting_line(line1,line2):
        if(line1.A*line2.B != line1.B*line2.A):
            return "Lines are intersecting"
        elif(line1.A*line2.B == line1.B*line2.A) and (line1.B*line2.C != line1.C*line2.B):
            return "Lines are parallel"
        elif(line1.A*line2.B == line1.B*line2.A) and (line1.B*line2.C == line1.C*line2.B):
            return "Lines are same or coincident"
```

```python
p1 = Point(0,0)
p2 = Point(1,10)
print(p1.distance_dtw_Two_Cordinate(p2))
# print(p2.distance_from_origin())
```

```python
l1 = Line(1,1,-2)
p1 = Point(1,1)
print(l1.point_on_line(p1))
```

```python
l1 = Line(1,1,-2)
p1 = Point(1,1)
print(l1.sortest_distance_point_line(p1))
```

```python
l1 = Line(2,3,1)
l2 = Line(4,5,2)
print(l1.intersecting_line(l2))
```
