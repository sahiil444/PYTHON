##### Aggregation (has a relationship)
         Aggregation is a relationship where one class contains an object of another class, but both can exist independently.

```python
class Customer:

    def __init__(self,name,gender,address):
        self.name = name
        self.gender = gender
        self.address = address

    def print_address(self):
        print(self.address.city,self.address.pincode,self.address.state)

    def edit_profile(self,new_name,new_city,new_pincode,new_state):
        self.name = new_name
        self.address.edit_address(new_city,new_pincode,new_state)
        

class Address:

    def __init__(self,city,pincode,state):
        self.city = city
        self.pincode = pincode
        self.state = state

    def edit_address(self,new_city,new_pincode,new_state):
        self.city = new_city
        self.pincode = new_pincode
        self.state = new_state

add1 = Address('Gola',262802,'UP')
cust1 = Customer('vikas','gay',add1)
```

```python
cust1.print_address()
cust1.edit_profile('Sahil','Lakhimpur kheri',262801,'utter pradesh')
cust1.print_address()
```

#### Inheritance (is a relationship):
        Inheritance is a relationship where one class inherits properties and methods of another class.
        ðŸ‘‰ Simple words:
                Parent â†’ Child
                Code reuse

```python
class Parents:

    def __init__(self):
        self.name = 'abdul sahil'
        self.gender = 'male'

    def login(self):
        print("login")

class Child(Parents):

    def enroll(self):
        self.roll = 10
```

```python
# Access of private attribute ?

class Phone:

    def __init__(self,brand,prize,camera):
        print("Inside the Phone (Parent class)")
        self.brand = brand
        self.__prize = prize
        self.camera = camera

    # To access the private attribute
    def get__Prize(self):
        return self.__prize

    
class Smartphone(Phone):

    def check(self):
        print(self.__prize)

s = Smartphone('IQOO',18000,64)
s.get__Prize()
```

#### Super() :
    super() is a built-in function used to call methods (especially the constructor) of the parent class from the child class.

#### Why do we need Super():
    Because when child class has its own __init__,
    parent ka constructor automatically run nahi hota âŒ

#### Conditions:
    super() can,t access attribute.
    super() cannot use outside the parent class
    super() always used inside the child class

```python
# (1)

class Phone:

    def __init__(self,brand,prize,camera):
        print("Inside the Phone (Parent class)")
        self.brand = brand
        self.prize = prize
        self.camera = camera

    def buy(self):
        print("Buying a phone")

    
class Smartphone(Phone):

    def buy(self):
        print("Buying a smartphone")
        super().buy()

s = Smartphone('IQOO',18000,64)
s.buy()
```

```python
# (2) Super --> Constructor

class Phone:

    def __init__(self,brand,prize,camera):
        print("Inside the Phone (Parent class)")
        self.brand = brand
        self.prize = prize
        self.camera = camera


    
class Smartphone(Phone):

    def __init__(self, brand, prize, camera,ram,os):
        print("inside child class (Smartphone)")
        super().__init__(brand, prize, camera)
        self.ram = ram
        self.os = os
        print("inside child class (Smartphone)")


s = Smartphone('IQOO',18000,64,8,'android')
print(s.brand)
print(s.prize)
```

```python
# (3) super() cannot use outside the parent class
#    super() cannot access attribute 

class Phone:

    def __init__(self,brand,prize,camera):
        print("Inside the Phone (Parent class)")
        self.brand = brand
        self.prize = prize
        self.camera = camera

    def buy(self):
        print("Buying a phone")

    
class Smartphone(Phone):

    def buy(self):
        print("Buying a smartphone")
        super().buy()

s = Smartphone('IQOO',18000,64)
# super().buy()
# print(s.super().brand)

```

#### ðŸ”¥ Types of Inheritance in Python

    Python supports 5 types:

       (1) Single Inheritance

       (2) Multilevel Inheritance

       (3) Multiple Inheritance

       (4) Hierarchical Inheritance

       (5) Hybrid Inheritance

```python
# Single Inheritance = One child class inherits from one parent class.

class Parent:
    def show(self):
        print("Parent method")

class Child(Parent):
    def display(self):
        print("Child method")

obj = Child()
obj.show()
obj.display()
```

```python
# (2) Multilevel Inheritance = A class inherits from a class which itself inherits from another class.

class A:
    def methodA(self):
        print("Class A")

class B(A):
    def methodB(self):
        print("Class B")

class C(B):
    def methodC(self):
        print("Class C")

obj = C()
obj.methodA()
obj.methodB()
obj.methodC()
```

```python
# (4) Hierarchical Inheritance = Multiple child classes inherit from one parent class.

class Parent:
    def show(self):
        print("Parent method")

class Child1(Parent):
    pass

class Child2(Parent):
    pass

obj1 = Child1()
obj2 = Child2()

obj1.show()
obj2.show()
```

```python
# (5) Multiple Inheritance = One child class inherits from more than one parent class.

class A:
    def methodA(self):
        print("Class A")

class B:
    def methodB(self):
        print("Class B")

class C(A, B):
    pass

obj = C()
obj.methodA()
obj.methodB()
```
