##### Functions On List

###### [1] list.append(element)  add element in end of list

```python
l1 = [1,3,6,2,7,3,9,4,8,5]
l1.append(10)
print(l1)
```

###### [2] list.insert(index,element) insert element at given index of list

```python
l1 = [1,3,6,2,7,3,9,4,8,5]
l1.insert(1,10)
print(l1)
```

###### [3] list.sort() sort in accending order

```python
l1 = [1,3,6,2,7,9,4,8,5]
l1.sort()
print(l1)
```

###### [4] list.sort(reverse=True) sort in deccending order

```python
l1 = [1,3,6,2,7,9,4,8,5]
l1.sort(reverse=True)
print(l1)
```

###### [5] list.reverse() reverse the whole list

```python
l1 = [1,3,6,2,7,9,4,8,5]
l1.reverse()
print(l1)
```

###### [6] list.remove(element)  remove first occurance of element 

```python
l1 = [1,3,6,3,2,7,3,9,4,8,5]
l1.remove(3)
print(l1)
```

###### [7] list.pop(index) remove element from list at the given index 

```python
l1 = [1,3,6,3,2,7,3,9,4,8,5]
l1.pop(6)
print(l1.pop(0))
print(l1)
```

```python
l1 = []  # empty list
l2 = [1]
l3 = [1,]
print(l1)
print(l2)
print(l3)
```

##### WAP to  find is given list is palindrom or not

```python
list = [1,2,3,2]
list2 = list.copy()
list2.reverse()
if(list == list2):
    print("Yes list is palindrom")
else:
    print("Not A palindrom")
```
