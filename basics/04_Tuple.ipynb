##### Creating Tuple 

```python
tup1 = () #empty tuple
print(tup1)


tup2  = (2,)
print(tup2)
print("Type of tup2:",type(tup2))

tup3 = (2)
print(tup3)
print("Type of tup3:",type(tup3))
```

##### Indexing And Selicing

```python
tup = (2,4,6,8,1,3,5,7,8)
print(tup[1:5])
```

##### Function in Tuple

```python
tup = (2,4,1,8,1,3,5,7,8)
# tup.index(element)  # return index of 1st occurance of elemenet
print(tup.index(1))
```

```python
tup = (2,4,1,8,1,3,1,7,8)
print(tup.count(1))  # count occurance of elemenet in tuple
```
