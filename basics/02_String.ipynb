##### Indexing In String And Selicing

```python
str = "Sahil Ansari"
print(str[4])
print(str[6])

print(str[6:12])
print(str[6:len(str)])
print(str[6:])  #str[6:len(str)]
print(str[:5])  #str[0:5]
print(str[-6:len(str)])
print(str[-6:-1])
```

##### Functions Of String

###### [1]  str.endswith("er")  return True if string ends with substr

```python
str = "hey i am learing python form apna collage"
print(str.endswith("age"))
print(str.endswith("e"))
print(str.endswith("a"))
```

######  [2]     str.find(word or latter)   return 1st index of 1st occur

```python
str = "hey i am learing python form apna collage"
print(str.find("python"))
print(str.find("h"))
print(str.find("Hate"))  # return -1 if no word is found in string
```

###### [3] str.capitalize()  capitalize fisrt latter of string

```python
str = "hey i am learing python form apna collage"
print(str.capitalize())
```

###### [4] str.count("latter OR word")   count occurance of word or latter or substract of string

```python
str = "hey i am learing from python from apna collage"
print(str.count("o"))
print(str.count("from"))
```

###### [5] str.replace(old value,new value)  replace all the occurance of old value with new value

```python
str = "hey i am learing from python from apna collage"
print(str.replace("from","By"))  # replace function create a new string thats why do not chnage in real string 
print(str)

# way by wih you can make change in string permanant not temporary
str = str.replace("from","By")
print(str)
```
