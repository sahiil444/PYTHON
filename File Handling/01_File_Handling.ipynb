#### Reading a file

```python
# reading from file using read()

f = open('sample.txt','r')
s = f.read()
print(s)
f.close()
```

```python
# reading upto given no. of character
f = open('sample.txt','r')
s = f.read(10)
print(s)
f.close()
```

```python
# readline()  --> read line by line 

f = open('sample.txt','r')
print(f.readline())
print(f.readline())
f.close()
```

```python
# print bhi line change karta hai and readline bhi line change krta hai thats why do line change hoge or ek line ka gape this can be solved by 

f = open('sample.txt','r')
print(f.readline(),end='')
print(f.readline(),end='')
f.close()
```

```python
# reading entire file using readline()
f = open('sample.txt','r')
while True:
    data = f.readline()
    if data != '':
        print(data,end='')
    else:
        break

f.close()
```

```python
# uning with
# with keyword closes the file as soon as usage is over 

with open('sample1.txt','w') as f:
    f.write('barketuuuuu')
```

#### serialization and deserialization 
    (1) serialization - Process of concerting python data type to JSON format
    (2) deserialization - Process of concerting JSON to python data format

```python
# serialization and deserialization in list
import json
l = [1,2,3,4]
with open('demo.json','w') as f:
    json.dump(l,f)

with open('demo.json','r') as f:
    data = json.load(f)
    print(data)
    print(type(data))
```

```python
# serialization and deserialization with dictionary
d = {
    'name':'sahil', 
    'branch':'CSE(AI/ML)', 
    'roll no': 202501100400010
}
with open('demo.json','w') as f:
    json.dump(d,f,indent=4)

with open('demo.json','r') as f:
    data = json.load(f)
    print(data)
    print(type(data))
```

```python
# serialization and deserialization with tuple 
t = (10,20,30,40,50)
with open('demo.json','w') as f:
    json.dump(t,f)

with open('demo.json','r') as f:
    data = json.load(f)
    print(data)
    print(type(data))

# Note : tuple ko as a list hi store karta hai
```

#### Serializing and Deserializing custom object

```python
class Person:

    def __init__(self,f_name,l_name,age,gender): 
        self.f_name = f_name
        self.l_name = l_name
        self.age = age
        self.gender = gender

person = Person('abdul','sahil',20,'male')
```

```python
import json
def format_type(person): 
    return f"{person.f_name} {person.l_name} age --> {person.age} gender --> {person.gender}"

with open('demo.json','w') as f:
    json.dump(person,f,default=format_type)
```

```python
with open('demo.json','r') as f:
    data = json.load(f)
    print(data)
    print(type(data))
```

```python
# print as a distionary
def format_type(person): 
    return {'name':person.f_name + ' ' + person.l_name,'age':person.age,'gender':person.gender}

with open('demo.json','w') as f:
    json.dump(person,f,default=format_type)
```

```python
with open('demo.json','r') as f:
    data = json.load(f)
    print(data)
    print(type(data))
```

#### Pickling and Unpickling

```python
class Person:

    def __init__(self,name,age,gender): 
        self.name = name
        self.age = age
        self.gender = gender

    def desiplay_info(self): 
        print("hi my name is",self.name,"and i am ",self.age,"year old")

person = Person('sahil',20,'male')
```
