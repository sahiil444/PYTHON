```python
#  catching specific exception
try:
    # m = 30
    f = open('new_sample.txt','r')
    print(f.read())
    print(m)
    # print(5/0)
    # print(p)

except FileNotFoundError:
    print('file not found')

except NameError:
    print('variable not define')

except ZeroDivisionError:
    print("can't divide by 0")

except Exception as e: 
    print(e)
    print(e.with_traceback)
```

```python
# else 
try:
    
    f = open('new_sample.txt','r')

except FileNotFoundError:
    print('file not found')

except NameError:
    print('variable not define')

except ZeroDivisionError:
    print("can't divide by 0")

except Exception as e: 
    print(e)
else:
    # vo part likhna hai jisme sure ho ki error nhi ayega 
    print(f.read())
```

```python
# finally

try:
    
    f = open('new_sample.txt','r')

except FileNotFoundError:
    print('file not found')

except NameError:
    print('variable not define')

except ZeroDivisionError:
    print("can't divide by 0")

except Exception as e: 
    print(e)
else:
    # vo part likhna hai jisme sure ho ki error nhi ayega 
    print(f.read())
finally:
    print("ye to har baar me print hoga chahe try run kre ya except ")
```

```python
class Bank:

    def __init__(self,balance): 
        self.balance = balance

    def withdraw(self,amount): 
        if amount<0: 
            raise Exception('amount cannot be negative')

        if amount>self.balance: 
            raise Exception('paise nhi hai apke pass')
        else:
            self.balance = self.balance-amount

obj = Bank(2000)
try:
    obj.withdraw(500)

except Exception as e:
    print(e)

else:
    print(obj.balance)

        
```
