#### Decorators:
    A decorator in python is a function that receives another function as input and adds some functionality(decoration) to and it and returns   it.
    This can happen only because python functions are 1st class citizens.
    There are 2 types of decorators available in python
    • Built in decorators like @staticmethod, @classmethod, @abstractmethod and @property etc
    • User defined decorators that we programmers can create according to our needs

```python
def my_decorator(func):
    def wrap():
        print("*************")
        func()
        print('*************')
    return wrap

# NOTE : decorator me parent function ki khatam hone ke baad bhi child function parent function ke variable and function ko access kr skta hai thats why
#        my_decorator ke khatam hone ke baad bhi func() ko print kr paa raha hai 

def hello():
    print('Hello')

a = my_decorator(hello)
a()
```

```python
def my_decorator(func):
    def wrap():
        print("*************")
        func()
        print('*************')
    return wrap


@ my_decorator
def hello():
    print('Hello')

hello()
```

```python
# use in some usefull
# calculate time taken by any function to run

import time
def timer(func):
    def wrap(*args):
        start = time.time()
        func(*args)
        print('time taken by ',func.__name__,time.time()-start,'secs')

    return wrap

@timer
def hello():
    time.sleep(2)
    print('hello world')

hello()

@timer
def power(a,b):
    print(a**b)

power(2,10)
```

```python
# checking input data type is valid or not 

def sanity_check(data_type):
    def outer_wrapper(func):
        def inner_wrapper(*args):
            if type(args[0]) == data_type:
                func(*args)
            else:
                raise TypeError('ye data type nhi chalega')

        return inner_wrapper
    return outer_wrapper


@sanity_check(int)
def square(num):
    print(num**2)


@sanity_check(str)
def hello(name):
    print('hello ',name)

square(2)
hello('sahil')
```
