### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
# scatter plot -> axes level function
sns.scatterplot(data=df,x='total_bill',y='tip')
```

```python
# relplot -> figure level -> square shape
sns.relplot(data=df,x='total_bill',y='tip',kind='scatter')
```

```python
# scatter plot -> axes level function
sns.scatterplot(data=df,x='total_bill',y='tip',hue='sex',size='size')
```

```python
# scatter using relplot -> size and hue
sns.relplot(data=df,x='total_bill',y='tip',kind='scatter',hue='sex',size='size')
```

```python
# style semantics
sns.relplot(data=df,x='total_bill',y='tip',kind='scatter',hue='sex',style='time')
```

```python
# line plot
gap = px.data.gapminder()
temp_df = gap[gap['country']=='India']
temp_df
```

```python
# axes level function
sns.lineplot(data=temp_df,x='year',y='lifeExp')
```

```python
# using relpplot
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

```python
# hue -> style->size
temp_df = gap[gap['country'].isin(['India','Pakistan','China'])]
sns.lineplot(data=temp_df,x='year',y='lifeExp',hue='country')
```

```python
# hue -> style->size
temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df.sample(5)
```

```python
# hue -> style using axes level plot
sns.lineplot(data=temp_df,x='year',y='lifeExp',hue='country',style='continent')
```

```python
# hue -> style using fig level plot means relplot
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line',hue='country',style='continent')
```

```python
df
```

```python
# facet plot only work with -> figure level function -> work with relplot
# it will not work with scatterplot and lineplot
sns.relplot(data=df,x='total_bill',y='tip',kind='scatter',col='sex')
```

```python
# row
sns.relplot(data=df,x='total_bill',y='tip',kind='scatter',row='sex')
```

```python
gap
```

```python
# col wrap
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year')
```

```python
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=3)
```

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

```python
# figure level -> displot
# axes level -> histplot -> kdeplot -> rugplot
```

```python
# plotting univariate histogram
sns.histplot(data=df,x='total_bill')
```

```python
sns.displot(data=df,x='total_bill',kind='hist')
```

```python
# bins parameter
sns.displot(data=df,x='total_bill',kind='hist',bins=20)
```

```python
# It’s also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorical variables

# countplot
sns.displot(data=df,x='day',kind='hist')
```

```python
# hue parameter
sns.displot(data=df,x='tip',kind='hist',hue='sex')
```

```python
# element -> step
sns.displot(data=df,x='tip',kind='hist',hue='sex',element='step')
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.displot(data=titanic,x='age',kind='hist',hue='sex',element='step')
```

```python
df
```

```python
# faceting using col and row -> not work on histplot function
sns.displot(data=df,x='tip',kind='hist',col='sex')
```

```python
df.head()
```

```python
# kdeplot : KDE = kernal density estimation
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate

sns.displot(data=df,x='total_bill',kind='kde')
```

```python
sns.kdeplot(data=df,x='total_bill')
```

```python
# hue -> fill
sns.displot(data=df,x='total_bill',kind='kde',hue='sex')
```

```python
sns.displot(data=df,x='total_bill',kind='kde',hue='sex',fill=True)
```

```python
# Rugplot

# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
sns.kdeplot(data=df,x='total_bill')
sns.rugplot(data=df,x='total_bill')
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot
# and then shows the count of observations within each rectangle with the fill color

# sns.histplot(data=tips, x='total_bill', y='tip')
sns.displot(data=df,x='total_bill',y='tip',kind='hist')
```

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
sns.kdeplot(data=df,x='total_bill',y='tip')
```
