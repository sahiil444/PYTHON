```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Colored Scatterplots
    A Colored Scatter Plot is a scatter plot in which data points are displayed with different colors to represent different categories or numerical values.

```python
iris = pd.read_csv('iris.csv')
iris
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'])
```

```python
# one dot represent one flower specise so i want to represent it in graph
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

    c parameter is used to specify the color of scatter plot point
                            OR
    The c parameter is used to assign colors to scatter plot points. It can take a single color, a list of colors, or numerical values for color mapping.

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'])
plt.colorbar()
```

    A colormap defines how numerical values are converted into colors.
    Popular colormaps:

        "viridis" ✅
        "plasma"
        "inferno"
        "cool"
        "hot"
        "jet"

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet')
plt.colorbar()
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

### Plot size

```python
plt.figure(figsize=(15,7))

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.9)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

### 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(8,4))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(8,4))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.log(xx)

fig = plt.figure(figsize=(8,4))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(6,3))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
