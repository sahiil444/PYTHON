#### 🔹 1. Pandas

Definition:
Pandas is a Python library used for data manipulation, analysis, and cleaning using structured data like tables.

#### 🔹 Series

Definition:
A Series is a one-dimensional labeled array that can store data of any type.

#### Series using list:

```python
# String

series = ['india','pakishtan','nepal','UAE','USA']
pd.Series(series)
```

```python
# Integer

runs = [56,23,70,88,65,100]
result = pd.Series(runs)

print(result)
```

```python
# Custom index

marks = [40,35,28,50]
subjects = ['Hindi','english','DS','LAE']

result = pd.Series(marks,index=subjects)
result
```

```python
# setting a name :


marks = [40,35,28,50]
subjects = ['Hindi','english','DS','LAE']

result = pd.Series(marks,index=subjects,name='Sahil ke marks:')
result
```

#### Series using dictionary

```python
marks = {
    'hindi':40,
    'english':32,
    'DS':22,
    'LAE':40
}

result = pd.Series(marks,name='Sahil ke marks')
result
```

#### Series Attribute:

```python

marks = {
    'hindi':40,
    'english':32,
    'DS':22,
    'LAE':40
}

runs = [34,56,78,89,90]

marks_series = pd.Series(marks,name='Sahil ke marks')
runs_series = pd.Series(runs,name="Kohli's Score")
```

```python
# Size : Returns the total number of elements in Series
print(marks_series.size)
print(runs_series.size)
```

```python
# dtype : Returns the data type of elements in the Series
print(marks_series.dtype)
print(runs_series.dtype)
```

```python
# name : Returns or sets the name of the Series
print(marks_series.name)
print(runs_series.name)
```

```python
# is_unique: returns True if series has unique items no duplicate items are present in series other wish returns False

print(marks_series.is_unique)
print(runs_series.is_unique)
print(pd.Series([1,1,2,3,4,5]).is_unique)
```

```python
# Index : Returns the index (labels) of the Series

print(marks_series.index)
print(runs_series.index)
```

```python
# Value:Returns the data of the Series as a NumPy array

print(marks_series.values)
print(runs_series.values)
```

#### 🔷 inplace Parameter
🔹 Definition (Exam Ready)

inplace is a parameter that decides whether the operation modifies the original data or returns a new object.

✅ 1. inplace = False (Default)

👉 Original data change nahi hota
👉 New object return hota hai

✅ 2. inplace = True ⭐

👉 Original data direct change ho jata hai
👉 Kuch return nahi hota (None)

```python
run.sort_values(inplace=True)
run
```

    🔷 4. drop_duplicates() ⭐

        Definition:
        drop_duplicates() removes duplicate rows from data.

```python
# duplicated() function gives boolean series .if value is duplicated than shows True otherwish False
print(duplicate_series)
print(duplicate_series.duplicated())
print(duplicate_series.duplicated().sum())
```
