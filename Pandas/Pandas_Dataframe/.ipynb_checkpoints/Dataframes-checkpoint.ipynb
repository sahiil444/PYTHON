### Selecting rows from a DataFrame

- **iloc** - searches using index positions
- **loc** - searches using index labels

```python
student_dict = {
    'name':['vikas','sahil','ashish','sumit','shzn'],
    'iq':[100,80,120,50,0],
    'marks':[90,70,120,40,-100],
    'package':[9,7,14,2,25]
}

students = pd.DataFrame(student_dict)
students.set_index('name',inplace=True) #set_index('column name jisse index bnana hai') : ye given column ko index bna deta hai
students
```

### Filtering a DataFrame

```python
ipl.head(2)
```

```python
# find all the final winners
new_def = ipl[ipl['MatchNumber'] == 'Final']
new_def[['Season','WinningTeam']]
```

```python
# how many super over finishes have occured
mask = ipl['SuperOver'] == 'Y'
new_def = ipl[mask]
print(new_def[['Team1','Team2','SuperOver']])

# ............................................................................................................................................

ipl[ipl['SuperOver'] == 'Y'].shape[0]
```

```python
# how many Final has csk won 
new_df = ipl[ipl['MatchNumber'] == 'Final']
new_df[new_df['WinningTeam'] == 'Chennai Super Kings'].shape[0]
```

```python
#  how many matches has csk won in kolkata

# new_df = ipl[ipl['WinningTeam'] == 'Chennai Super Kings']
# new_df[new_df['City'] == 'Kolkata'].shape[0]

ipl[(ipl['WinningTeam'] == 'Chennai Super Kings') & (ipl['City'] == 'Kolkata')].shape[0]
```

```python
# toss winner is match winner in percentage
x = ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0]
total_match = ipl.shape[0]
print((x*100)/total_match)
```

```python
# movies with rating higher than 8 and votes>10000

movie = movies[(movies['imdb_rating']>8) & (movies['imdb_votes']>10000)]
movie[['title_y','original_title','imdb_rating','imdb_votes']]
```

```python
# Action movies with rating higher than 7.5

# method1
# movies[(movies['genres'].str.split('|').apply(lambda x : 'Action' in x)) & (movies['imdb_rating']>7.5)]

# ............................................................................................................................................
# method2

movies[(movies['genres'].str.contains('Action')) & (movies['imdb_rating']>7.5)]
```
