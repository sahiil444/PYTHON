    ### Matplotlib
        Matplotlib is a Python library used for data visualization by plotting graphs and charts

    ### Types of Data

        1: Numerical Data
            📌 Definition:

        Numerical data is the type of data that consists of numbers and can be measured or counted.

        🔹 Examples:
        Age = 18, 20, 25
        Marks = 85, 90
        Height = 170 cm

        🔹 Types of Numerical Data:

        Discrete Data → Countable
        👉 Example: Number of students, cars

        Continuous Data → Measurable
        👉 Example: Height, weight, temperature


        2: Categorical Data

            📌 Definition:

        Categorical data is the type of data that represents categories or labels instead of numbers.

        🔹 Examples:
        Gender = Male, Female
        Color = Red, Blue
        Grade = A, B, C

        🔹 Types of Categorical Data:
        Nominal Data → No order
        👉 Example: Colors, names
        
        Ordinal Data → Ordered
        👉 Example: Rank (1st, 2nd), Grades (A, B, C)

    color (line ka color)
        ✔ Options:

```python
# colors(hex) and line(width and style) and marker(size)

plt.plot(batsman['index'],batsman['V Kohli'],color='pink')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#3C8753')

plt.title('Rohit VS Virat IPL career Compresion')
plt.xlabel('Session')
plt.ylabel('Runs')
```

### marker (points ka shape)


    linestyle (line ka style):
         Types:

        '-' → solid
        '--' → dashed
        ':' → dotted
        '-.' → dash-dot


    linewidth (line ki thickness)

```python
# marker:

plt.plot(batsman['index'],batsman['V Kohli'],color='pink',linestyle='solid',linewidth=2,marker='D',markersize=5)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#3C8753',linestyle='dotted',linewidth=2,marker='o',markersize=5)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

    label (legend ke liye):
        label tab tak show nhi hoga jab tK plt.legend() nhi run kroge 
        and by-default ye best jagah pe graph me show hoga 

```python
# legend -> location

plt.plot(batsman['index'],batsman['V Kohli'],color='pink',linestyle='solid',linewidth=2,marker='D',markersize=5,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#3C8753',linestyle='dotted',linewidth=2,marker='o',markersize=5,label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend()
```

    Limiting axes means setting the range of values displayed on the X-axis and Y-axis in a graph.

        👉 It is used to:

        Zoom into a specific part of data
        Remove unnecessary area
        Improve visualization


        xlim() → X-axis control
        ylim() → Y-axis control

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0,75000)
# plt.xlim(2017,2019)
```

    Grid:  is a set of horizontal and vertical lines in a graph that helps in reading values more clearly.

        👉 It improves:

        Readability
        Data interpretation
        Alignment of points

```python
# grid

plt.plot(batsman['index'],batsman['V Kohli'],color='pink',linestyle='solid',linewidth=2,marker='D',markersize=5)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#3C8753',linestyle='dotted',linewidth=2,marker='o',markersize=5)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()
```

    plt.show():  is a function used to display the plotted graph on the screen.

    🔹 Why it is needed?

    👉 Without plt.show():

    Graph may not appear (especially in scripts)
    Plot remains in memory only

    👉 With plt.show():

    Graph window opens
    Output becomes visible

```python
# show
plt.plot(batsman['index'],batsman['V Kohli'],color='pink',linestyle='solid',linewidth=2,marker='D',markersize=5)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#3C8753',linestyle='dotted',linewidth=2,marker='o',markersize=5)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()

plt.show()
```

plt.plot vs plt.scatter

| Feature       | `plt.plot(...,'o')` | `plt.scatter()`         |
| ------------- | ------------------- | ----------------------- |
| Purpose       | Line plot           | Scatter plot            |
| Color control | Limited             | Advanced                |
| Size control  | Same for all        | Different size possible |
| Color mapping | ❌                   | ✅ (`c=`)                |
| Bubble chart  | ❌                   | ✅                       |

    plt.xticks():
        X-axis ke numbers/labels ko control karta hai

    Syntex:
        plt.xticks(ticks, labels)
        ticks → positions
        labels → names shown on axis

        ex:
        plt.xticks([1,2,3], ['A','B','C'])
        X-axis me 1,2,3 ki jagah A,B,C dikhega

    ...........................................................

        plt.xticks(rotation=45)
        Labels ko tilt (ghuma) deta hai 45 degree se 

```python
# xticks
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='pink')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='green')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
# dataset
df = pd.read_csv('gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
